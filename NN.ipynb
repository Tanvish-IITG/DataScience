```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, Y_train, Y_test = train_test_split(digits.data,digits.target,test_size = 0.8)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
img = X_test[1].reshape(8,8)
plt.matshow(img)
```

```python
Y_train[1]
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(64,64,64,64,64),activation = 'relu',solver = 'lbfgs',max_iter=500,alpha=1)
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict_proba([X_test[1]])
```

```python
from sklearn.svm import SVC
model1 = SVC(kernel = 'linear')
model1.fit(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
plt.figure(figsize = (20,20))
sn.heatmap(cm,annot = True)
```

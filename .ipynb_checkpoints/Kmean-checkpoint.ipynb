```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df
```

```python
iris.target
```

```python
iris.target_names
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
Setosa = df.loc[df['target'] == 0,['petal length (cm)' ,'petal width (cm)']]
Setosa.head()
```

```python
Virsicolor = df.loc[df['target'] == 1,['petal length (cm)' ,'petal width (cm)']]
Virsicolor.head()
```

```python
Virginia = df.loc[df['target'] == 2,['petal length (cm)' ,'petal width (cm)']]
Virginia.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(Setosa['petal length (cm)'],Setosa['petal width (cm)'],color = 'b')
plt.scatter(Virsicolor['petal length (cm)'],Virsicolor['petal width (cm)'],color = 'orange')
plt.scatter(Virginia['petal length (cm)'],Virginia['petal width (cm)'],color = 'g')
```

```python
Data = df.loc[:,['petal length (cm)' ,'petal width (cm)']]
```

```python
Data
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
pred = model.fit_predict(Data)
```

```python
pred
```

```python
Data['pred'] = pred
```

```python
Data.head()
```

```python
plt.scatter(Data['petal length (cm)'][Data.pred == 2],Data['petal width (cm)'][Data.pred == 2],color = 'b')
plt.scatter(Data['petal length (cm)'][Data.pred == 1],Data['petal width (cm)'][Data.pred == 1],color = 'orange')
plt.scatter(Data['petal length (cm)'][Data.pred == 0],Data['petal width (cm)'][Data.pred == 0],color = 'g')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker = '*',color = 'black')
plt.xlabel('petal length')
plt.ylabel('sepal length')
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
Rng = range(1,11)
l = []
for i in Rng:
    model = KMeans(n_clusters=i)
    model.fit_predict(Data)
    l.append(model.inertia_)
l
```

```python
plt.scatter(Rng,l)
plt.plot(Rng,l)
```
